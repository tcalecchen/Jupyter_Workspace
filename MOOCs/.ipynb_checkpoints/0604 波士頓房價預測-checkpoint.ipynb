## `scikit-learn` 真實世界數據

`scikit-learn` 內建一些真實世界的數據, 可以讓你玩玩, 他們稱做 "Toy Datasets"。有哪些可以參考 [scikit-learn 官網](http://scikit-learn.org/stable/datasets/index.html)的說明。

````python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
````

#### 讀入 boston 房價數據

````python
from sklearn.datasets import load_boston
````

````python
boston = load_boston()
````

## 資料裡到底有什麼

#### features

你可以用 `feature_names` 看到數據中所有的 features。你才發現原來有 13 個 features!

````python
boston.feature_names
````

````python
X = boston.data
Y = boston.target  # 將士頓房地產資料的預測數值匯入到 Y
print(X)
print('===========================================================================')
print(Y)
````

````python
len(X)
````

````python
x_train, x_test, y_train, y_test = train_test_split(X, Y,
                                                   test_size=0.3,
                                                   random_state=87)
````

````python
x_train.shape
````

````python
regr = LinearRegression()
````

````python
regr.fit(x_train, y_train)
````

````python
y_predict = regr.predict(x_test)
````

````python
plt.scatter(y_test, y_predict)
#plt.plot(x_test, y_predict, 'r')
plt.plot([0,50],[0,50],'r')
plt.xlabel('True Price')
plt.ylabel('Predicted Price')
````

### [小技巧] 善用 `enumerate`

#### eunmerate()是一個內建的函式，用於在一個可迭代或可遍歷的物件（如列表、字串、陣列、字典）中，將物件組成一個一個序列和索引，可以同時獲得索引和索引值。

````python
L = ['a', 'b', 'c']
````

````python
for i in L:
    print(i)
````

````python
for i in range(3):
    print(i+1, L[i])
````

````python
list(enumerate(L))
````

### enumerate(list 變數)，enumerate 會把 list 的 index 與 list 的 item 包成一個 tuple，然後再把每個 tuple 包成按 index 順序的 tuple。

````python
for i in enumerate(L):
    print(i)
````

````python
for i, s in enumerate(L):
    print(i+1, s)
````

### [小技巧] 畫多個圖

````python
x = np.linspace(-10,10,200)
````

````python
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
````

````python
plt.subplot(2,2,1)      # 2 x 2 第一張圖
plt.plot(x, np.sin(x))

plt.subplot(2,2,2)      # 2 x 2 第二張圖
plt.plot(x, np.cos(x))

plt.subplot(2,2,3)      # 2 x 2 第三張圖
plt.plot(x, x)

plt.subplot(2,2,4)      # 2 x 2 第四張圖
plt.plot(x, x**2)
````

#### 炫炫的畫出個別參數和 target 關係



````python
plt.figure(figsize=(8,10))  # figsize，指定 figure 的寬和高，單位為英寸
for i, feature in enumerate(boston.feature_names):
    plt.subplot(5, 3, i+1)
    plt.scatter(X[:,i], Y, s=1)  # X = boston.data, X[:,i] 所有的列，第 i 行的資料， Y = boston.target
    plt.ylabel("price")
    plt.xlabel(feature)
    plt.tight_layout()  # 會自動調整子圖參數，使之填充整個圖像區域
````
