```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.random.rand(100,2)  # 隨機產生 100 列, 2 行的資料
```

```python
X.shape
```

```python
X[0]
```

```python
plt.scatter(X[:,0], X[:,1], s=50)
```

```python
from sklearn.cluster import KMeans  # 匯入KMeans函式庫
```

```python
clf = KMeans(n_clusters=3)  # 開 K-Mean 機器的方法，機器變數名 = KMeans(n_clusters=你要分幾群)，這裡分成 3 群
```

```python
clf.fit(X)  # 開始訓練
```

```python
clf.labels_  # 取得預測結果，已經分成 3 群（0, 1, 2）
```

```python
plt.scatter(X[:,0], X[:,1], c=clf.labels_)  # c=clf.labels_ 分群結果
```
