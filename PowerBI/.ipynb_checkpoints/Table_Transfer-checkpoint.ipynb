```python
import pandas as pd
```

```python
# 讀取資料
data_path = 'data/'
df = pd.read_excel(data_path + '寬表格轉長表格練習資料.xlsx')
print(df)
```

```python
# 建立一個空的DataFrame
df_all = pd.DataFrame()
```

```python
# 每一個寬表格佔用 66列，所以每 66 列來對表格進行切分，切分後轉換成長表格，再合併到 df_all
for i in range(int(df.shape[0]/66)):
    # 一次選擇 66 列，並重設 index
    df_part = df.iloc[66*i:66*(i+1), :].reset_index(drop=True)

    # 需要選取的列
    selected_rows = [0, 2] + list(range(4, 66))

    # 篩選列、重設 index、並轉置表格
    df_part = df_part.iloc[selected_rows, :].reset_index(drop=True).T

    # 重新命名欄位
    df_part.columns = ['居住地', '年齡層'] + df_part.iloc[0, 2:].tolist()

    # 篩選掉不需要的列
    df_part = df_part.iloc[1:, :]

    # 向下填補缺漏值
    df_part['居住地'] = df_part['居住地'].fillna(method='ffill')

    # 由寬表格轉換成長表格
    df_part_long = pd.melt(df_part, id_vars=df_part.columns[0:2], value_vars=df_part.columns[2:], var_name='年份', value_name='來臺人次')

    # append 到 df_all
    df_all = pd.concat([df_all, df_part_long])
    
    print(df_all)
```
