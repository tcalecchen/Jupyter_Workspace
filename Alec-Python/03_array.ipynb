```python
A = np.random.randn(100)  # 平均值 0，標準差 1 的一維矩陣
```

```python
A
```

```python
A = 10*A + 50  # 標準差 10，平均值 50 的一維矩陣
```

```python
A
```

```python
A.shape    # 矩陣 A 的維度，一維：100、二維：0
```

```python
A.shape = (5, 20) # 將矩陣 A 轉變成一維：5、二維：20 的二維矩陣，一維與二維相乘必須是 100
```

```python
A
```

```python
A.reshape(100,1)  # 會在記憶體內產生一個新的 Array，不會改變原來的 A 
```

```python
A  # reshape 後 A 還是沒變
```

```python
# 如果要改變 A，可以重新指派

A = A.reshape(100,1)  
```

```python
A   # 重新指派後，A 就改變了
```
