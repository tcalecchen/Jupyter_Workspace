### 長條圖

```python
#!/usr/bin/env python3
import matplotlib.pyplot as plt
plt.style.use('ggplot')  # 使用 ggplot 樣式表來模仿 ggplot2 的風格，這是一種受歡迎的 R 繪圖套件。

# 建立長條圖的資料
customers = ['John', 'Mary', 'Andy', 'Ryan', 'Lucy']
customers_index = range(len(customers))  # 建立一串顧客索引值，因為 xticks 函式需要用索引位置與標籤來設定標籤。
print("customers_index = ", customers_index)
sale_amounts = [127, 90, 201, 111, 232]

fig = plt.figure()  # 建立一張圖。
ax1 = fig.add_subplot(1,1,1)  # 在圖內加入一張小圖，1,1,1 代表一列，一欄，以及第一個且唯一的小圖。
ax1.bar(customers_index, sale_amounts, align='center', color='darkorange')  # 建立長條圖，customers_index 指定長條圖左側
# 的 x 座標，sale_amounts 指定長條的高度，align='center'指定長條在它們的標籤上置中，color='darkblue'指定長條顏色。
ax1.xaxis.set_ticks_position('bottom') # 指定刻度標記位於下方，移除上方標記。
ax1.yaxis.set_ticks_position('left')   # 指定刻度標記位於左方，移除右方標記。
plt.xticks(customers_index, customers, rotation=30, fontsize='small') # 將長條的刻度標籤文字從顧客的索引號碼改為它們的姓名
# rotation=30 指定刻度標籤文字的角度。fontsize='small' 會減少刻度標籤文字的大小。

plt.xlabel('Customer Name')  # x 軸標籤。
plt.ylabel('Sale Amount')    # y 軸標籤。
plt.title('Sale Amount per Customer') # 圖表的標題。

plt.savefig('bar_plot.png', dpi=400, bbox_inches='tight') # 在目前資料夾將圖表存為 bar_plot.png，bbox_inches='tight' 在儲存
# 的圖表周圍加上空白區域。
plt.show()  # 在螢幕新視窗顯示圖表。
```

### 色階分布圖

```python
#!/usr/bin/env python3
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')

mu1, mu2, sigma = 100, 130, 15
# 使用亂數產生器來建立兩個常態分布變數。
x1 = mu1 + sigma*np.random.randn(10000)
x2 = mu2 + sigma*np.random.randn(10000)

fig = plt.figure()
ax1 = fig.add_subplot(1,1,1)
n, bins, patches = ax1.hist(x1, bins=50, density=False, color='darkgreen') # density=False 表示這個色階分布圖應顯示頻率分布
                                                                           # 而不是機率分布。
n, bins, patches = ax1.hist(x2, bins=50, density=False, color='orange', alpha=0.5) # alpha=0.5 表示比較透明。
ax1.xaxis.set_ticks_position('bottom')
ax1.yaxis.set_ticks_position('left')

plt.xlabel('Bins')
plt.ylabel('Number of Values in Bin')
fig.suptitle('Histograms', fontsize=14, fontweight='bold') # 為圖表加入一個置中標題，字體大小 14，粗體。
ax1.set_title('Two Frequency Distributions') # 為小圖加入置中標題，在圖表的標題下面。

plt.savefig('histogram.png', dpi=400, bbox_inches='tight')
plt.show()
```
