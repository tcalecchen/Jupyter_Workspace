```python
df = pd.read_csv("http://bit.ly/uforeports")
```

```python
df.head(10)
```

```python
df_state = df.groupby("State").count()
```

```python
df_state
```

```python
df_state.sort_values(by="Time", ascending=False)  # 以 Time 欄位的值做升冪排序
```

```python
df_state  # df_state 做完 sort 後不會改變原來的 dataframe 內容，而是產生一個新的 dataframe
```

```python
df_state.sort_values(by="Time", ascending=False, inplace=True) # 如果要改變原來 dataframe，要下 inplace=True
```

```python
df_state.head(10)
```

```python
df_state[:10].Time.plot(kind='bar')
```
