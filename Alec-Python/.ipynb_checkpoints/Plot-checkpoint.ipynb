## 長條圖

### 長條圖代表的是許多種類的數值，例如數量。常見的長條圖包括垂直與水平圖、堆疊圖及分組圖。

```python
#!/usr/bin/env python3
import matplotlib.pyplot as plt
plt.style.use('ggplot')  # 使用 ggplot 樣式表來模仿 ggplot2 的風格，這是一種受歡迎的 R 繪圖套件。

# 建立長條圖的資料
customers = ['John', 'Mary', 'Andy', 'Ryan', 'Lucy']
customers_index = range(len(customers))  # 建立一串顧客索引值，因為 xticks 函式需要用索引位置與標籤來設定標籤。
print("customers_index = ", customers_index)
sale_amounts = [127, 90, 201, 111, 232]

fig = plt.figure()  # 建立一張圖。
ax1 = fig.add_subplot(1,1,1)  # 在圖內加入一張小圖，1,1,1 代表一列，一欄，以及第一個且唯一的小圖。
ax1.bar(customers_index, sale_amounts, align='center', color='darkorange')  # 建立長條圖，customers_index 指定長條圖左側
# 的 x 座標，sale_amounts 指定長條的高度，align='center'指定長條在它們的標籤上置中，color='darkblue'指定長條顏色。
ax1.xaxis.set_ticks_position('bottom') # 指定刻度標記位於下方，移除上方標記。
ax1.yaxis.set_ticks_position('left')   # 指定刻度標記位於左方，移除右方標記。
plt.xticks(customers_index, customers, rotation=30, fontsize='small') # 將長條的刻度標籤文字從顧客的索引號碼改為它們的姓名
# rotation=30 指定刻度標籤文字的角度。fontsize='small' 會減少刻度標籤文字的大小。

plt.xlabel('Customer Name')  # x 軸標籤。
plt.ylabel('Sale Amount')    # y 軸標籤。
plt.title('Sale Amount per Customer') # 圖表的標題。

plt.savefig('bar_plot.png', dpi=400, bbox_inches='tight') # 在目前資料夾將圖表存為 bar_plot.png，bbox_inches='tight' 在儲存
# 的圖表周圍加上空白區域。
plt.show()  # 在螢幕新視窗顯示圖表。
```

## 色階分布圖

### 色階分布圖可表示數值的分布情形。常見的色階分布圖包括頻率分布、頻率密度分布、機率分布，以及機率密度分布。

```python
#!/usr/bin/env python3
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')

mu1, mu2, sigma = 100, 130, 15
# 使用亂數產生器來建立兩個常態分布變數。
x1 = mu1 + sigma*np.random.randn(10000)
x2 = mu2 + sigma*np.random.randn(10000)

fig = plt.figure()
ax1 = fig.add_subplot(1,1,1)
n, bins, patches = ax1.hist(x1, bins=50, density=False, color='darkgreen') # density=False 表示這個色階分布圖應顯示頻率分布
                                                                           # 而不是機率分布。
n, bins, patches = ax1.hist(x2, bins=50, density=False, color='darkorange', alpha=0.5) # alpha=0.5 表示比較透明。
ax1.xaxis.set_ticks_position('bottom')
ax1.yaxis.set_ticks_position('left')

plt.xlabel('Bins')
plt.ylabel('Number of Values in Bin')
fig.suptitle('Histograms', fontsize=14, fontweight='bold') # 為圖表加入一個置中標題，字體大小 14，粗體。
ax1.set_title('Two Frequency Distributions') # 為小圖加入置中標題，在圖表的標題下面。

plt.savefig('histogram.png', dpi=400, bbox_inches='tight')
plt.show()
```

## 線型圖

### 線型圖可呈現沿著一條數字線的數值。它經常被用來顯示資料隨著時間推移的情形。

```python
#!/usr/bin/env python3
from numpy.random import randn
import matplotlib.pyplot as plt
plt.style.use('ggplot')

# 用 randn 建立要繪製的隨機資料
plot_data1 = randn(50).cumsum()
plot_data2 = randn(50).cumsum()
plot_data3 = randn(50).cumsum()
plot_data4 = randn(50).cumsum()

fig = plt.figure()
ax1 = fig.add_subplot(1,1,1)
ax1.plot(plot_data1, marker=r'o', color=u'blue', linestyle='-', label='Blue Solid')
ax1.plot(plot_data2, marker=r'+', color=u'red', linestyle='--', label='Red Dashed')
ax1.plot(plot_data3, marker=r'*', color=u'green', linestyle='-.', label='Green Dash Dot')
ax1.plot(plot_data4, marker=r's', color=u'orange', linestyle=':', label='Orange Dotted')
ax1.xaxis.set_ticks_position('bottom')
ax1.yaxis.set_ticks_position('left')

ax1.set_title('Line Plots: Markers, Colors, and Linestyles')
plt.xlabel('Draw')
plt.ylabel('Random Number')
plt.legend(loc='best')  # 建立圖表的圖例，loc='best' 會指示 matplotlib 將圖例放在圖表開放空間的最佳位置

plt.savefig('line_plot.png', dpi=400, bbox_inches='tight')
plt.show()
```

## 散布圖

### 散布圖可展示兩個座標軸所代表的數值變數的值，例如，身高 vs. 體重，或供應 vs. 需求。散布圖可提供一些提示，讓你瞭解變數是否呈現正相關（即，都集中在一個特定的位置）或負相關（即，點會像擴散雲一樣散開）。你也可以繪製穿越點的迴歸線、最小平方誤差，來根據一個變數的值來預測另一個變數。

```python
#!/usr/bin/env python3
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')

x = np.arange(start=1., stop=15., step=1.)
# 使用稍微偏離線性與二次多項式方程式的亂數來建立資料
y_linear = x + 5. * np.random.randn(14)
y_quadratic = x**2 + 10. * np.random.randn(14)

# 使用 numpy 的 polyfit 與 polyld 函式來建立通過 (x, y_linear) 與 (x, y_quadratic) 的線性直線與二次多項式方程式曲線。
# 在真實世界的資料中，你可以使用 polyfit 函式來計算多項式的係數，來指定擬合程度。polyld 函式使用係數來建立實際的多項式方程式。
fn_linear = np.poly1d(np.polyfit(x, y_linear, deg=1))
fn_quadratic = np.poly1d(np.polyfit(x, y_quadratic, deg=2))

fig = plt.figure()
ax1 = fig.add_subplot(1,1,1)

# 建立有兩條迴歸線的散布圖，'bo' 代表 (x, y_linear) 點是藍色圓圈，'go' 代表 (x, y_quadratic) 點是綠色圓圈。同樣的，'b-'
# 代表穿越 (x, y_linear) 點的線條是藍色實線，'g-' 代表穿越 (x, y_quadratic) 點的線條是綠色實線。可以用 linewidth 來指定線寬。
ax1.plot(x, y_linear, 'bo', x, y_quadratic, 'go', \
			x, fn_linear(x), 'b-', x, fn_quadratic(x), 'g-', linewidth=2.)
ax1.xaxis.set_ticks_position('bottom')
ax1.yaxis.set_ticks_position('left')

ax1.set_title('Scatter Plots with Best Fit Lines')
plt.xlabel('x')
plt.ylabel('f(x)')

# 設定 x 軸與 y 軸的限制。線條使用 min 與 max 函式，根據實際的資料值來限制座標軸。你也可以使用特定的數字，例如 xlim(0, 20) 與
# ylim(0, 200)。如果你沒有指定座標軸限制，matplotlib 會為你設定限制。
plt.xlim((min(x)-1., max(x)+1.))
plt.ylim((min(y_quadratic)-10., max(y_quadratic)+10.))

plt.savefig('scatter_plot.png', dpi=400, bbox_inches='tight')
plt.show()
```

## 盒鬚圖

### 盒鬚圖會根據資料的最小值、第一四分位數、中位數、第三四分位數，以及最大值來顯示資料。盒子的最下面與最上面是第一與第三四分位數，盒子中間那條線是中位數。從盒子兩端延伸的直線（稱為鬚）是最小與最大的非離群值，而鬚之外的點是離群值。

```python
#!/usr/bin/env python3
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')

N = 500
normal = np.random.normal(loc=0.0, scale=1.0, size=N)
lognormal = np.random.lognormal(mean=0.0, sigma=1.0, size=N)
#index_value = np.random.random_integers(low=0, high=N-1, size=N)  # This function is deprecated
index_value = np.random.randint(low=0, high=N-1, size=N)
normal_sample = normal[index_value]
lognormal_sample = lognormal[index_value]
box_plot_data = [normal,normal_sample,lognormal,lognormal_sample]

fig = plt.figure()
ax1 = fig.add_subplot(1,1,1)

# 建立一個串列 box_labels，裡面存有每一個盒鬚圖的標籤。下一行的 boxplot 函式會使用這個串列
box_labels = ['normal','normal_sample','lognormal','lognormal_sample']

# 使用 boxplot 函式來建立四個盒鬚圖。 notch=False 代表這些箱子應該都是矩形，而不是在中間有缺口。 sym='.' 代表鬚之外的點是圓點，
# 而不是預設的 + 符號。 vert=True 代表盒子是垂直的，而不是水平的。 whis=1.5 指定鬚接觸的地方在第一四分位數與第三四分位數之外
# （即，Q3 + whis*IQR，IQR = 四分位範圍，Q3-Q1）。 showmeans=True 指定盒子除了顯示中位值之外，也要顯示平均值。
# labels=box_labels 代表使用 box_labels 的值來標示盒鬚圖。
ax1.boxplot(box_plot_data, notch=False, sym='.', vert=True, whis=1.5, \
				showmeans=True, labels=box_labels)
ax1.xaxis.set_ticks_position('bottom')
ax1.yaxis.set_ticks_position('left')
ax1.set_title('Box Plots: Resampling of Two Distributions')
ax1.set_xlabel('Distribution')
ax1.set_ylabel('Value')

plt.savefig('box_plot.png', dpi=400, bbox_inches='tight')
plt.show()
```

## pandas

### pandas 藉由提供可操作 Series 與 DataFrame 的 plot 函式，來簡化以 Series 與 DataFrame 內的資料繪製圖表的程序。在預設情況下，plot 函式會建立線形圖，但你可以使用 kind 引數來建立不同類型的圖。
### 例如，除了標準的統計圖之外，你可以用 matplotlib 來繪圖，pandas 可讓你建立其他的圖表，例如六角相形圖、散布矩陣圖、密度圖、Andrews 曲線、平行座標、滯後圖、自相關圖，與 bootstrap 圖。pandas 也可讓你輕鬆地在圖中加入輔助 y 軸、誤差線及運算列表。


```python
#!/usr/bin/env python3
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')

fig, axes = plt.subplots(nrows=1, ncols=2)  # 
ax1, ax2 = axes.ravel()

data_frame = pd.DataFrame(np.random.rand(5, 3),
						index=['Customer 1', 'Customer 2', 'Customer 3', 'Customer 4', 'Customer 5'],
						columns=pd.Index(['Metric 1', 'Metric 2', 'Metric 3'], name='Metrics'))

data_frame.plot(kind='bar', ax=ax1, alpha=0.75, title='Bar Plot')
plt.setp(ax1.get_xticklabels(), rotation=45, fontsize=10)
plt.setp(ax1.get_yticklabels(), rotation=0, fontsize=10)
ax1.set_xlabel('Customer')
ax1.set_ylabel('Value')
ax1.xaxis.set_ticks_position('bottom')
ax1.yaxis.set_ticks_position('left')

colors = dict(boxes='DarkBlue', whiskers='Gray', medians='Red', caps='Black')
data_frame.plot(kind='box', color=colors, sym='r.', ax=ax2, title='Box Plot')
plt.setp(ax2.get_xticklabels(), rotation=45, fontsize=10)
plt.setp(ax2.get_yticklabels(), rotation=0, fontsize=10)
ax2.set_xlabel('Metric')
ax2.set_ylabel('Value')
ax2.xaxis.set_ticks_position('bottom')
ax2.yaxis.set_ticks_position('left')

plt.savefig('pandas_plots.png', dpi=400, bbox_inches='tight')
plt.show()
```
